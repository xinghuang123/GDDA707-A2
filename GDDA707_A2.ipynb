```python
# Part A, Task 2, a, Load the datasets

import pandas as pd

df_2009 = pd.read_csv('online_retail_2009.csv')
df_2010 = pd.read_csv('online_retail_2010.csv')

df_2009.head()
```

```python
df_2010.head()
```

```python
# Part A, Task 2, b, Merging
df_combined = pd.concat([df_2009, df_2010], ignore_index=True)
df_combined.head()
```

```python
df_combined.info()
```

```python
df_combined.describe()
```

```python
# Part A, Task  2, b, Cleaning
df_combined.shape
```

```python
# Part A, Task  2, b, Cleaning
# Drop rows without "CustomerID"
df_combined.dropna(subset=['CustomerID'], inplace=True)
```

```python
# Drop rows with negative "Quantity" and "Price"
df_combined = df_combined[(df_combined['Quantity'] > 0) & (df_combined['Price'] > 0)]
```

```python
# Handling outliers in the 'Price' column using the IQR method

# Calculating Q1 (25th percentile) and Q3 (75th percentile) for 'Price'
Q1_price = df_combined['Price'].quantile(0.25)
Q3_price = df_combined['Price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

# Filtering out the outliers in the 'Price' column
df_combined_no_outliers_price = df_combined[
    (df_combined['Price'] >= (Q1_price - 1.5 * IQR_price)) & 
    (df_combined['Price'] <= (Q3_price + 1.5 * IQR_price))
]

# Displaying the first few rows of the DataFrame without outliers in 'Price'
df_combined_no_outliers_price.head()
```

```python
df_combined.shape
```

```python
df_combined_no_outliers_price.shape
```

```python
# Part A, Task 2, b, Formatting

# Converting the 'CustomerID' column to string type
df_combined_no_outliers_price['CustomerID'] = df_combined_no_outliers_price['CustomerID'].astype(str)

# Verifying the conversion
df_combined_no_outliers_price['CustomerID'].dtype
```

```python
# Part A, Task 2, b, Formatting

# Converting the 'InvoiceDate' column to datetime type with dayfirst=True
df_combined_no_outliers_price['InvoiceDate'] = pd.to_datetime(
    df_combined_no_outliers_price['InvoiceDate'], 
    dayfirst=True
)

# Verifying the conversion
df_combined_no_outliers_price['InvoiceDate'].dtype
```

```python
# Step 2

# Create a database and a collection
db = client['online_retail']
collection = db['online_retail_collection']

# Insert the filtered data into the collection
collection.insert_many(df_combined.to_dict('records'))
```
